```python
## Phase 1: Basics of Matplotlib
# Topic 1: What is Data Visualization?
# Data Visualization means representing data using charts and graphs.
```

```python
# Line chart
import matplotlib.pyplot as plt
months = ["Jan","Feb","Mar","Apr","May"]
sales = [10,20,40,60,80]


plt.figure(figsize=(8,9))
plt.plot(months,sales)
plt.show()
```

```python
# Chart Title and Labels

plt.plot(months, sales)

plt.title("monthly sales")
plt.xlabel("Months")
plt.ylabel("sales")
plt.show()
```

```python
# Change Color and Style
plt.plot(months,sales,color = "red",linestyle="--",marker = "*")
```

```python
# Bar chart
import matplotlib.pyplot as plt
students = ["Ashok","Anusha","bablu"]
marks = [90,70,50]

plt.barh(students,marks)
```

```python
import matplotlib.pyplot as plt
students = ["Ashok","Anusha","bablu"]
marks = [90,70,50]

plt.bar(students,marks)
```

```python
import matplotlib.pyplot as plt
subject = ["java","Sql","Springboot"]
hours = [20,12,60]

plt.pie(hours,labels = subject,autopct = "%1.1f%%")
```

```python
# Scatter plot
import matplotlib.pyplot as plt
age = [20,25,30,35,40]
salary = [25000,30000,40000,50000,60000]

plt.scatter(age,salary)
plt.show()
```

```python
# Phase 2: Intermediate Matplotlib
# Figure Size
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr"]
sales = [10,20,40,60]

plt.figure(figsize=(8,5))

plt.plot(months,sales)

plt.show()


# common sizes
# plt.figure(figsize=(5,3))
# plt.figure(figsize=(8,5))
# plt.figure(figsize=(10,6))
# plt.figure(figsize=(12,7))
```

```python
# Grid
import matplotlib.pyplot as plt
students = ["Ashok","Paatu","Ace"]
marks = [12,3,55]

plt.plot(students,marks)
plt.grid(True)
plt.show()
```
