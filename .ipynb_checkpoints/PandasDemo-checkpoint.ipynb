```python
# Pandas Day-2
# Agenda : How to handle data?

# 1. Data selection
# 2. Handling missing values
# 3. Datatype conversion
# 4. Removing duplicates
```

```python
## 1. Data Selection
# row and column selection
df.loc[0:7,"Region":"Sales_Channel"]      # Name wise selection
```

```python
df.iloc[0:2,0:5]                            #Index wise selection
```

```python
## 2. Handling missing values

#Checking null values
df.isnull().sum() # each column null value count
```

```python
# Dropping null values 
df.dropna()
```

```python
# Filling null values
df.fillna(0)
```

```python
## 3. Data Type Conversions
df['Units_Sold'] = df['Units_Sold'].astype('float')
```

```python
df.info()
```

```python
df.head()
```

```python
##4. Removing duplicates rows
df.drop_duplicates()
```

```python
# Pandas Day-3
# Agenda : Data Transformation & Analysis
                        
# 1. Sorting data
# 2. Value counts
# 3. grouping data
# 4. Feature Engineering
# 5. Pivot table
```

```python
# 1. Sorting data
df.sort_values(by = 'Unit_SellingPrice',ascending = False)
```

```python
df.sort_values(by = ["Region","Country"], ascending = [False,True])
```

```python
# 2. Value counts
df["Region"].value_counts()
```

```python
# 3. grouping data
df.groupby(['Sales_Channel'])["Total_Revenue"].mean()
```

```python
# 4. Feature Engineering
df["Channel_Type"] = df["Sales_Channel"]

# Display result
df[['Sales_Channel','Channel_Type']].head()
```
